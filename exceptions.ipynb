```python
def divide(n1,n2):
    print(f"number 1={n1}\nnumber 2={n2}")
    result=0
    try:
        result=n1/n2
    except ZeroDivisionError:
        print(f"number 2 {n2} is zero, you cannot divide by zero.")
    else:

        print(f"result of dividing {n1} by {n2} is {result}")
    return result
```

```python
for i in range(10):
    divide(i+1,i)
```
