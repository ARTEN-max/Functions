```python
def loan_emi(amount,duration,rate,down_payment=0):
  """Calculates the equal monthly installment (EMI) for a loan.

    Arguments:
          amount - Total amount to be spent (loan + down payment)
          duration - duration of the loan
          rate- rate of interest
          down payment - optional initial payment
  """
  loan_amount=amount-down_payment
  try:

    emi = loan_amount*rate*((1+rate)**duration)/(((1+rate)**duration)-1)

  except ZeroDivisionError:
    emi=loan_amount/duration

    emi=math.ceil(emi)

  return emi
```

```python
Paris=[200,20,200,"Paris"]
London=[250,30,120,"London"]
Dubai=[370,15,80,"Dubai"]
Mumbai=[420,10,70,"Mumbai"]
Cities=[Paris,London,Dubai,Mumbai]
```

```python
def cost_of_trip(flight,hotel_cost,car_rent,number_of_days=0):
  return flight+(hotel_cost*number_of_days)+(car_rent*math.ceil(number_of_days)/7)
```

```python

def days_to_visit(days):
  costs=[]
  for city in Cities:
    cost=cost_of_trip(city[0],city[1],city[2],days)
    cost.append((cost,city[3]))
    min_cost=min(costs)
    return min_cost
```
