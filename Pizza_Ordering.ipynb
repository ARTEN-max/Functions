```python
"""
*ask customer to select size of pizza and base
*ask customer for choice of toppings
*reask customer to confirm choices,alter or stop
*if customer asks to alter regive choices of what he want to alter
*if customer stops stop the order
*if customer confirms order produce a random unique order number for the customer
"""
```

```python
#*Dictionary
#this dictionary is used to get the most and least popular toppings.
popularToppings={
    "Pepperoni":0,
    "Chicken":0,
    "Extra Cheese":0,
    "Mushrooms":0,
    "Spinach":0,
    "Olives":0
}
#this dictionary stores the user choices like size base and toppings
userOrder={
    "size":"",
    "base":"",
    "toppings":["","",""]

}
#this list holds the order numbers of all the orders, is used to get the total number of orders
orderHistory=[]
```

```python
#*ask customer to select size of pizza
def getPizzaSize():
  size=0
  while size<=0 or size>3:
    size=int(input("Choose your pizza size(1.Small,2.Medium,3.Large)"))
    if size==1:
      return("small")
    if size==2:
      return("medium")
    if size==3:
      return("large")


```

```python
def getPizzaBase():
  """
  ask customer to select base of pizza
  """
  base=0
  while base<=0 or base>2:
    base=int(input("Choose your pizza base(1.Thin,2.Thick)"))
    if base==1:
      return("thin")
    if base==2:
      return("thick")

```

```python

def getUserToppings():
  """
  ask customer for choice of toppings
  """
  count = 0

  toppings = []
  while count < 3:
    choice = input('Do you need any additional toppings: (Y/N)')
    if choice == 'y':
      topping = getPizzaToppings()
      toppings.append(topping)
    elif choice == 'n':
      break
    count = count + 1
  print(f'The toppings selected = {toppings}')
  return toppings




```

```python

def getPizzaToppings():
  """
  You should provide a menu of choices for toppings to be selected from the user.
  """
  menu = ['Pepperoni','Chicken','Extra Cheese','Mushrooms','Spinach','Olives']
  choice = 0
  while choice <=0 or choice > 6:
    print('Chose the toppings from below: ')
    print("1.Pepperoni\n2.Chicken\n3.Extra Cheese\n4.Mushrooms\n5.Spinach\n6.Olives")
    choice = int(input())
  return menu[choice-1]
```

```python
def displayUserOrder():
  """
  Diplays order to user in form of list
  """
  global userOrder
  print(f'chosen size={userOrder["size"]}')
  print(f'chosen base={userOrder["base"]}')
  print(f'chosen toppings={userOrder["toppings"]}')
```

```python
import random

def confirmOrder():
  """
  reask customer to confirm choices,alter or stop
  """
  global userOrder
  #get user choice for pizza size
  userOrder["size"]=getPizzaSize()
  #get user choice for pizza base
  userOrder["base"]=getPizzaBase()
  #get the additional topping list selected from the user
  toppingsList=getUserToppings()
  #update the topping list dictionary
  userOrder["toppings"]=toppingsList
  #display the order and take user confirmation
  displayUserOrder()
  orderConfirmation=input("would you like to confirm your order (Y/N)")
  #if the user confirms the order generate a order number and also update the popular toppings dictionary
  if orderConfirmation=="y":
    orderNumber=random.randint(1,1000)
    orderHistory.append(orderNumber)
    for topping in toppingsList:
      popularToppings[topping]=popularToppings[topping]+1
    return orderNumber
  #give user the option to alter his/her choices,once done display the order and return the order number
  else:
    sizeChoice=input("would you like to alter the size of the pizza (Y/N)")
    if sizeChoice=="y":
      userOrder["size"]=getPizzaSize()
    baseChoice=input("would you like to alter the base of the pizza (Y/N)")
    if baseChoice=="y":
      userOrder["base"]=getPizzaBase()
    toppingsChoice=input("would you like to alter the toppings of the pizza (Y/N)")
    if toppingsChoice=="y":
      toppingsList.clear()
      toppingsList=getUserToppings()
      userOrder["toppings"]=toppingsList
  displayUserOrder()
  orderNumber=random.randint(1,1000)
  orderHistory.append(orderNumber)
  for topping in toppingsList:
    popularToppings[topping]=popularToppings[topping]+1
  return orderNumber


```

```python
def getMostPopularToppings():
  global popularToppings
  mostPopularTopping=max(popularToppings,key=popularToppings.get)
  return mostPopularTopping,popularToppings[mostPopularTopping]
getMostPopularToppings()

```

```python
def getLeastPopularToppings():
  global popularToppings
  leastPopularTopping=min(popularToppings,key=popularToppings.get)
  return leastPopularTopping,popularToppings[leastPopularTopping]
getLeastPopularToppings()
```
